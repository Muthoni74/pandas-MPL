```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 2]

# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Sample Data')

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Matplotlib Plot Example')

# Add grid for better readability
plt.grid(True)

plt.legend()

# Save the plot to a file
plt.savefig('simple_matplotlib_plot.png') # Added this line to save the plot

# Show the plot
plt.show()
```

This code generates a simple line plot with markers, using `matplotlib.pyplot`. It defines `x` and `y` coordinates, then uses `plt.plot()` to create the line. Labels, a title, and a grid are added for clarity before `plt.show()` displays the plot.
