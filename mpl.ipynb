**Reasoning**:
To visualize the distribution of the 'Attrition' column, I will create a bar chart showing the counts of 'Yes' and 'No' attrition cases. This will provide an immediate understanding of the class balance for the target variable.



```python
plt.figure(figsize=(6, 4))
df['Attrition'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Attrition')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Reasoning**:
To visualize the distribution of employees across different 'Department' categories, I will create a bar chart showing the counts for each department. This will help understand the departmental composition of the workforce.



```python
plt.figure(figsize=(8, 5))
df['Department'].value_counts().plot(kind='bar', color=['lightgreen', 'cornflowerblue', 'lightcoral'])
plt.title('Distribution of Employees Across Departments')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
